```python
def plot_trace(trace_nr, segfile):
    dt=segfile[0].stats.segy.trace_header.sample_interval_in_ms_for_this_trace/1000

    ntraces = len(segfile.traces)
    nsamples = len(segfile.traces[trace_nr].data)
    
    times= np.linspace(0, dt*(nsamples-1), nsamples)

    print (len(times))

    fig= plt.figure(1, figsize=(12, 4))
    plt.plot(times, segfile[trace_nr].data, label='Amplitude')

    plt.grid(True)
    plt.legend(loc='upper right')
    trace_name='Trace '+str(trace_nr)
    plt.title(trace_name)
    plt.show()
```

### Make df for each sgy file

```python
def make_df(sgyfile, column_name):
    start_time = time.time()
    xline=[]
    inline=[]
    dt=[]
    Traces=[]
    for tr in range (0, len(sgyfile)):
        Traces.append (sgyfile[tr].data)
        inline.append (sgyfile[tr].stats.segy.trace_header.source_coordinate_x)
        xline.append (sgyfile[tr].stats.segy.trace_header.source_coordinate_y)
        dt.append(sgyfile[tr].stats.segy.trace_header.sample_interval_in_ms_for_this_trace/1000)
        
    print (set(dt))
    
    # make multi index
    arrays=[inline, xline]
    tuples = list(zip(*arrays))
    index = pd.MultiIndex.from_tuples(tuples, names=["Inline", "Xline"])
    
    Traces_df=pd.DataFrame(Traces)
    Traces_df.index=index
    
    df=pd.DataFrame(Traces_df.stack(), columns=[column_name])
    df.reset_index(inplace=True)
    df.rename(columns={'level_2':'Time'}, inplace=True)
    df['Time']=df['Time']*4
    
    print("--- %s seconds ---" % (time.time() - start_time))
    
    return df
```

```python
density_df=make_df(segyfile1, 'Density')
pwave_df=make_df(segyfile2, 'P_wave')
vclay_df=make_df(segyfile4, 'V_clay')
```

```python
porosity_df=make_df(segyfile3, 'Porosity')
```

```python
density_df.shape, pwave_df.shape, porosity_df.shape, vclay_df.shape
```

```python
# here i use concat with assumptions that all inline, xline and time are the same and ordered right way
# there was a problem with porositty inline/xline
# it's better to use merge (make sure that xline/inline are the same) 

start_time = time.time()
d3_df=pd.concat([density_df, pwave_df['P_wave']], axis=1)
d3_df=pd.concat([d3_df, porosity_df['Porosity']], axis=1)
d3_df=pd.concat([d3_df, vclay_df['V_clay']], axis=1)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
start_time = time.time()
d3=pd.merge(density_df, pwave_df, on=['Inline', 'Xline', 'Time'], how='outer')
d3=pd.merge(d3, porosity_df, on=['Inline', 'Xline', 'Time'], how='outer')
d3=pd.merge(d3, vclay_df, on=['Inline', 'Xline', 'Time'], how='outer')
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
d3_1=d3[~d3['Density'].isna()]
d3_1.drop('Porosity', axis=1, inplace=True)

d3_1_df=d3_df.drop('Porosity', axis=1)

d3_1.equals(d3_1_df)
```
